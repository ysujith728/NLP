# Exercise 2: Demonstration of the Phases of NLP using NLTK

This notebook demonstrates the 5 main phases of Natural Language Processing (NLP) using Python's NLTK library, running on the contents of `food.txt`.

## The 5 Phases of Natural Language Processing

1. **Lexical Analysis (Tokenization):** Breaking down the text into paragraphs, sentences, and words.
2. **Syntactic Analysis (Parsing / POS Tagging):** Checking grammar and word relationships in sentences.
3. **Semantic Analysis:** Determining the literal meaning of words and phrases (e.g., entity identification, dictionary definitions).
4. **Discourse Integration:** Understanding the meaning of sentences in relation to other sentences (e.g., coreference resolution like resolving pronouns).
5. **Pragmatic Analysis:** Interpreting the text in context to extract the real-world intent, attitude, or sentiment.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag, ne_chunk
from nltk.corpus import wordnet
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import os
import ssl

# Handle SSL certificate issues when downloading resources in some Windows environments
try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

# Download the required NLTK resources (includes both legacy and new NLTK package names)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('maxent_ne_chunker', quiet=True)
nltk.download('words', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('vader_lexicon', quiet=True)
```

```python
# Read the food.txt file (handles running from folder or workspace root)
file_name = 'food.txt'
possible_paths = [
    file_name,
    os.path.join('Lab1-Introduction-to-NLTK', file_name)
]

text = None
for path in possible_paths:
    if os.path.exists(path):
        with open(path, 'r', encoding='utf-8') as f:
            text = f.read()
        break

if text is None:
    raise FileNotFoundError(f"Could not find {file_name} in the expected paths.")

print("=== Input Text ===")
print(text)
```

```python
# Phase 1: Lexical Analysis (Tokenization & Segmentation)
sentences = sent_tokenize(text)
tokens = word_tokenize(text)

print("=== Phase 1: Lexical Analysis ===")
print(f"Total sentences segmented: {len(sentences)}")
print(f"First sentence: {sentences[0]}")
print(f"Total word tokens extracted: {len(tokens)}")
print(f"First 10 tokens: {tokens[:10]}")
```

```python
# Phase 2: Syntactic Analysis (POS Tagging)
# Annotate the first sentence with Part-of-Speech tags
pos_tags = pos_tag(word_tokenize(sentences[0]))

print("=== Phase 2: Syntactic Analysis ===")
print(f"POS Tags for: '{sentences[0]}'")
for word, tag in pos_tags:
    print(f"  {word:<15} -> {tag}")
```

```python
# Phase 3: Semantic Analysis (Word Meaning & Entity Recognition)
print("=== Phase 3: Semantic Analysis ===")

# 1. Find synonyms/definitions using WordNet
synsets = wordnet.synsets("food")
if synsets:
    print("WordNet definition of 'food':")
    print(f"  {synsets[0].definition()}")

# 2. Identify Named Entities in a sentence
sample_sentence = "Sujith loves eating Mutton Biryani."
entities = ne_chunk(pos_tag(word_tokenize(sample_sentence)))
print(f"\nNamed Entity Recognition on '{sample_sentence}':")
for chunk in entities:
    if hasattr(chunk, 'label'):
        print(f"  {chunk.label()}: {' '.join(c[0] for c in chunk)}")
```

```python
# Phase 4: Discourse Integration (Context / Reference Resolution)
print("=== Phase 4: Discourse Integration ===")
print("Comparing consecutive sentences:")
print(f"  Sentence 1: '{sentences[0]}'")
print(f"  Sentence 2: '{sentences[1]}'")
print("\nDiscourse Resolution:")
print("  In sentence 2, the pronoun 'its' is a coreference pointing to the entity 'mutton biryani' introduced in sentence 1.")
```

```python
# Phase 5: Pragmatic Analysis (Sentiment & Intent Analysis)
sia = SentimentIntensityAnalyzer()
overall_sentiment = sia.polarity_scores(text)
conclusion_sentence = sentences[-1]
conclusion_sentiment = sia.polarity_scores(conclusion_sentence)

print("=== Phase 5: Pragmatic Analysis ===")
print("Overall text sentiment analysis:", overall_sentiment)
print(f"\nFoodie conclusion sentence: '{conclusion_sentence}'")
print("Conclusion sentence sentiment score:", conclusion_sentiment)
```
