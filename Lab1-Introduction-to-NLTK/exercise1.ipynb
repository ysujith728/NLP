# Exercise 1: Tokenization and Stop Words using NLTK

This notebook loads the text from `myself.txt`, tokenizes it, identifies the stop words present, and prints the sentences with stop words removed using Python's NLTK library.

```python
import os
import string
import ssl
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords

# Handle SSL certificate issues when downloading resources in some Windows environments
try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

# Download the required NLTK resource packages quietly to keep outputs clean
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
```

```python
# Load text from myself.txt (handles running from folder or workspace root)
file_name = 'myself.txt'
possible_paths = [
    file_name,
    os.path.join('Lab1-Introduction-to-NLTK', file_name)
]

text = None
for path in possible_paths:
    if os.path.exists(path):
        with open(path, 'r', encoding='utf-8') as file:
            text = file.read()
        break

if text is None:
    raise FileNotFoundError(f"Could not find {file_name} in the expected paths.")

print("=== Original Text ===")
print(text)
```

```python
# Step 1: Tokenize the text into words
tokens = word_tokenize(text)

print("=== Word Tokens ===")
print(tokens)
```

```python
# Step 2: Identify and print stop words
stop_words = set(stopwords.words('english'))

# Extract stop words present in the text
stop_words_in_text = [word for word in tokens if word.lower() in stop_words]
unique_stop_words = sorted(list(set(w.lower() for w in stop_words_in_text)))

print("=== Stop Words in Text ===")
print(f"All Stop Words Found ({len(stop_words_in_text)}): {stop_words_in_text}")
print(f"Unique Stop Words Found ({len(unique_stop_words)}): {unique_stop_words}")
```

```python
# Step 3: Segment into sentences and print them without stop words
sentences = sent_tokenize(text)

print("=== Sentences Without Stop Words ===")
for i, sentence in enumerate(sentences, 1):
    words_in_sentence = word_tokenize(sentence)
    # Filter out stop words
    filtered_words = [word for word in words_in_sentence if word.lower() not in stop_words]
    
    # Join words and clean up spacing around punctuation marks
    filtered_sentence = " ".join(filtered_words)
    for punct in string.punctuation:
        filtered_sentence = filtered_sentence.replace(" " + punct, punct)
        
    print(f"Sentence {i} (Original): {sentence.strip()}")
    print(f"Sentence {i} (Filtered): {filtered_sentence}\n")
```
